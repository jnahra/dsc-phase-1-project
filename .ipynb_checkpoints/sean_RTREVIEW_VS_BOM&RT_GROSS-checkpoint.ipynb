Noticed that many of the reviews are either strings such as B+ or C, or a division such as 2/5 or 1/10 or 1/4. 

Will replace values within each of their own categories to give them a new review grade of:

A 100-90% , B 80-89% , C 70-79% , D 60-69%, F 0-59%

A is any high rating at or above a 90% score
B is any middle rating such as 80-89%
C is any low-middle rating such as 70-79%
D is any low rating from 60-69%
F is rating below a 59% and lower


Will sort by the top critic reviews. 

```python
#Deleting the characters of '-' and '+' from column rating

rt['rating'] = rt['rating'].str.replace('-', '').str.replace('+', '')
```

Will be changing data to search through top_critic value only ratings

```python
# select rows where top_critic column is equal to 1
rt_top = rt[rt['top_critic'] == 1]

# view the selected rows
rt_top.head()
```

```python
#View the number of null values in new dataframe rt_top
rt_top['rating'].isna().sum()
```

```python
replace_dict = {"A": 95, "B": 85, "C": 75, "D": 65, "F": 0}
rt_top = rt_top.assign(rating=rt_top["rating"].replace(replace_dict))

```

```python
rt_top[20:30]
```

```python
from fractions import Fraction

# run all fractions and create their result as the new value within the cell


# Define a function to convert fractions to decimal values
def convert_fraction(fraction):
    if pd.isna(fraction):
        return None
    elif isinstance(fraction, str):
        if '/' in fraction:
            numerator, denominator = fraction.split('/')
            return float(numerator)/float(denominator) * 100
        else:
            return float(fraction)
    else:
        return fraction

# Apply the function to the "rating" column
rt_top["rating"] = rt_top["rating"].apply(convert_fraction)
```

```python
rt_top[20:30]
```

```python
rt_top['rating'].head(20)
```

```python
# Compute the mean of the "rating" column
mean_rating = round(rt_top["rating"].mean())

# Replace all NaN values in the "rating" column with the mean
rt_top["rating"].fillna(mean_rating, inplace=True)
```

Following code changes values of each non-numerical rating to relate to the column.

Changing rating values to integers to lead to final analysis

```python
rt_top = rt_top.rename(columns={'rating': 'rating (%)'})
```
