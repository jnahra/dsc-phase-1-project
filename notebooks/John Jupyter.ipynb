```python
df_budget = pd.read_csv('zippedData/tn.movie_budgets.csv.gz')
```

```python
df_budget.head()
```

```python
df_budget.tail()
```

```python
df_budget['worldwide_gross'] = df_budget['worldwide_gross'].str[1:]
```

```python
df_budget['worldwide_gross'] = df_budget['worldwide_gross'].str.replace(',','')
```

```python
df_budget['worldwide_gross'] = df_budget['worldwide_gross'].astype(int)
```

```python
df_budget['production_budget'] = df_budget['production_budget'].str[1:]
```

```python
df_budget['production_budget'] = df_budget['production_budget'].str.replace(',','')
```

```python
df_budget['production_budget'] = df_budget['production_budget'].astype(int)
```

```python
df_budget['profit'] = df_budget['worldwide_gross'] - df_budget['production_budget']
```

```python
df_budget.sort_values('profit',ascending=False)
```

```python
df_budget['ROI'] = df_budget['profit'] / df_budget['production_budget'] * 100
```

```python
df_budget.sort_values('profit',ascending=False)[0:50]
```

```python
df_budget.sort_values('ROI',ascending=False)[0:50]
```
